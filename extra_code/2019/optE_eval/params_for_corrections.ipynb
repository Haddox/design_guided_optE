# Make a list of parameters for the corrections file in Rosetta

Read in parameters from flags file

```python
flags_f = 'data/HH_run25B_flags_311_final'
with open(flags_f) as f:
    all_lines = f.readlines()
```

set_atom_properties

```python
# Get entries with '-chemical:set_atom_properties' and
# then remove that flag, leaving the value
lines = sorted([
    line.strip().replace(
        '-chemical:set_atom_properties ', ''
    ) for line in all_lines
    if '-chemical:set_atom_properties' in line
])

# Get data for specific property and insert value into
# a line for the params file
prop = ':'
lines = [
    f'\t\tparams.push_back("{line}");' for line in lines
    if prop in line
]
print('\n'.join(lines))
```
